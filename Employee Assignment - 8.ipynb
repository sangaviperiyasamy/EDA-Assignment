```python
#The sns.set_style sets the aesthetics of the plot 
#%config InlineBackend figure_format = ‘retina’  makes the plot higher resolution. 
#%matplotlib inline is a magic function that tells Matplotlib to generate our plots within the frontend 
import pandas as pan
import numpy as npy
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
sns.set_style('darkgrid')
%config InlineBackend.figure_format = 'retina'
plt.style.use('ggplot')
```

```python
emp = pan.read_csv("WA_Fn-UseC_-HR-Employee-Attrition.csv")
emp.head()
```

```python
#cleaning the dataset
#removing the unnecessary columns
del_col = ['Attrition','BusinessTravel','DailyRate','DistanceFromHome','EnvironmentSatisfaction','HourlyRate',
           'JobInvolvement','JobLevel','JobSatisfaction','MonthlyRate','NumCompaniesWorked','Over18','OverTime',
           'RelationshipSatisfaction','StockOptionLevel','WorkLifeBalance','YearsWithCurrManager','YearsSinceLastPromotion',
          'TrainingTimesLastYear','Education','EducationField','MaritalStatus','YearsInCurrentRole','EmployeeCount',
           'StandardHours']

emp = emp.drop(del_col, axis=1)
emp.head()
```

```python
#  dropping the duplicates from the dataset.
print(emp.shape)
emp = emp.drop_duplicates(keep='first')
print(emp.shape)
```

```python
#final dataset looks like
emp.head(10)
```

```python
#1.changing the name of the column index
column_name = emp.columns.values
column_name[2] = 'Employee_Id'
```

```python
emp.head()
```

```python
#2.changing the row index name
row_name = emp.index.values
row_name[1] = '1111'
```

```python
emp.head(15)
```

```python
#3.sort the rows of dataframe in Place based on row index label names
emp.sort_index(inplace=True)
print('Contents of Row Index Labels are :')
print(emp)
```

```python
# 4.sort a dataframe based on column names
col = emp.sort_index(axis=1)
print('Contents of Column Names are :')
print(col)
```

```python
#5.Pass the row elements as key value pairs to append() function 
row_app = emp.append({'Age' : 23 , 'Department' : 'IT','Employee_Id' : 3456,'Gender' : 'Female','JobRole':'Software Engineer',
                        'MonthlyIncome' : 7000,'PercentSalaryHike' : 12,'PerformanceRating' : 5,
                        'TotalWorkingYears' : 7 ,'YearsAtCompany' : 2} , ignore_index=True)
row_app
```

```python
#6.Pass a series in append() to append a row in dataframe  
rows_app = emp.append(pan.Series([ 21, 'IT',7890,'Female','Software engineer',10000,10,4,8,5], 
                                index=emp.columns ), ignore_index=True)
rows_app
```

```python
#data visualization
#plotting in line chart
emp.plot(x = 'PercentSalaryHike', y = 'PerformanceRating', kind='line')
```

```python
#plotting in pie chart
emp.plot(kind = 'line', y='PerformanceRating', figsize=(8,6),color = 'blue')
plt.title('Ratings')
plt.show()
```

```python
#plotting in box chart
emp.plot(kind='box', figsize=(15, 8))
plt.title('Box plot of employee')
plt.ylabel('Employee details')
plt.show()
```

```python
#Plotting in bar chart with two-level group by
emp.groupby(['PercentSalaryHike','Gender']).size().unstack().plot(kind='bar',stacked=True)
plt.show()
```

```python
#plotting in bar chart
emp.groupby(['Gender','Department']).size().unstack().plot(kind='bar',stacked=True)
plt.show()
```
