```python
#The sns.set_style sets the aesthetics of the plot 
#%config InlineBackend figure_format = ‘retina’  makes the plot higher resolution. 
#%matplotlib inline is a magic function that tells Matplotlib to generate our plots within the frontend 
import pandas as pan
import numpy as npy
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
sns.set_style('darkgrid')
%config InlineBackend.figure_format = 'retina'
```

```python
#reading csv file
fire = pan.read_csv('forest_fires.csv')
fire.head(10)
```

```python
#cleaning the dataset
# removing duplicates from the row
print(fire.shape)
fire= fire.drop_duplicates(keep='first')
print(fire.shape)
```

```python
#replacing all the zeros from columns
cols = ['X','Y','FFMC','DMC','DC','ISI','temp','RH','wind','rain','area']
fire[cols] = fire[cols].replace(0, npy.nan)
fire.head(15)
```

```python
#1. Displaying the highest DMC by month wise in descending order
fire = fire.sort_values(by =['month', 'DMC'], ascending = False)
fire.head(10)
```

```python
#2.converting a dataframe column into a list
list_of_mon = fire['month'].to_list()
print(list_of_mon)
type(list_of_mon)
```

```python
#3.finding the minimum temp in the list
minValue = fire['temp'].min()
print(minValue)
```

```python
#4.# Get frequency count of values in column DC
frequency = fire['DC'].value_counts()
print(frequency.head(15))
```

```python
fire.tail(5)
```

```python
#5.displaying columns in reverse order

for column in reversed(fire.columns):
   # Select column contents by column name using [] operator
   reverse_fire = fire[column]
   print(column, '\n')
   print( reverse_fire.tail(),'\n')
print("\t\t\t\t\n")
```

```python
#6.count all nan in dataframe in both rows and columns
fire.isnull().sum().sum()
```

```python
#data visualization
# displaying in scatter plot
fire.plot(kind='scatter',x='month',y='day',color='red')
plt.show()
```

```python
#displaying in bar chart
fire.plot(kind='bar',x='FFMC',y='DMC',color = 'blue')
```
