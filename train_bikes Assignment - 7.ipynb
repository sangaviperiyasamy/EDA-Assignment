```python
#The sns.set_style sets the aesthetics of the plot 
#%config InlineBackend figure_format = ‘retina’  makes the plot higher resolution. 
#%matplotlib inline is a magic function that tells Matplotlib to generate our plots within the frontend 
import pandas as pan
import numpy as npy
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
sns.set_style('darkgrid')
%config InlineBackend.figure_format = 'retina'
```

```python
# Reading the csv file
train_bike = pan.read_csv('train_bikes.csv')
train_bike.head(10)
```

```python
#cleaning the dataset
# replacing all the zeros from holiday , workingday and windspeed cols.
cols = ['holiday', 'windspeed','workingday']
train_bike[cols] = train_bike[cols].replace(0, npy.nan)
train_bike.head()
```

```python
#1.Getting a column name from dataframe
columns_name = train_bike.columns.values
columns_name
```

```python
#2.Getting a column index label name from dataframe
train_bike.columns.values[2]
```

```python
#3.It returns an ndarray of all row indexes in dataframe
row_names = train_bike.index.values
row_names
```

```python
#4.Get Row Index name by position in DataFrame
train_bike.index.values[5]
```

```python
#5.printing a specified column values
print(train_bike['season'])
```

```python
#data visualization
#correlation matrices using heatmap
plt.subplots(figsize = (15,15))
sns.heatmap(train_bike.corr(),annot = True)
```

```python
#plotting in histogram
train_bike["season"].hist()
```

```python
#plotting in line chart
train_bike.plot.line(x='casual', y='registered', figsize=(6,4))
plt.show()
```

```python
#plotting in boxplot chart
train_bike.plot.box(figsize=(8,6))
plt.show()
```

```python
#plotting hexagonal chart in pandas
train_bike.plot.hexbin(x='count', y='season',gridsize=20, figsize=(8,10))
```

```python
#plotting in kernal density chart
train_bike["humidity"].plot.kde()
```
